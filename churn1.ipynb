```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter

# Load the dataset
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Convert TotalCharges to numeric (empty strings will become NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fill NaN values in TotalCharges with 0 (for new customers)
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Convert Churn to binary (1 for Yes, 0 for No)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Set style for plots
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# 1. Churn Distribution
plt.subplot(2, 2, 1)
churn_counts = df['Churn'].value_counts()
plt.pie(churn_counts, labels=['No Churn', 'Churn'], autopct='%1.1f%%', 
        colors=['#66b3ff','#ff9999'], startangle=90)
plt.title('Overall Churn Rate')

# 2. Churn by Contract Type
plt.subplot(2, 2, 2)
contract_churn = df.groupby('Contract')['Churn'].mean().sort_values()
contract_churn.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)

# 3. Churn by Internet Service
plt.subplot(2, 2, 3)
internet_churn = df.groupby('InternetService')['Churn'].mean().sort_values()
internet_churn.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Churn Rate by Internet Service')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)

# 4. Tenure Distribution by Churn
plt.subplot(2, 2, 4)
sns.boxplot(x='Churn', y='tenure', data=df, palette=['#66b3ff','#ff9999'])
plt.title('Tenure Distribution by Churn Status')
plt.xlabel('Churn')
plt.ylabel('Tenure (months)')
plt.xticks([0, 1], ['No', 'Yes'])

plt.tight_layout()
plt.show()

# 5. Monthly Charges vs Total Charges by Churn
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MonthlyCharges', y='TotalCharges', hue='Churn', 
                data=df, palette=['#66b3ff','#ff9999'], alpha=0.6)
plt.title('Monthly Charges vs Total Charges by Churn Status')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Total Charges ($)')
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.show()

# 6. Payment Method Impact on Churn
plt.figure(figsize=(12, 6))
payment_churn = df.groupby('PaymentMethod')['Churn'].mean().sort_values(ascending=False)
payment_churn.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Churn Rate by Payment Method')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
plt.show()

# 7. Correlation Heatmap of Numerical Features
plt.figure(figsize=(10, 6))
numeric_df = df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn', 'SeniorCitizen']]
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# 8. Distribution of Monthly Charges by Churn
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', 
            palette=['#66b3ff','#ff9999'], fill=True, common_norm=False)
plt.title('Distribution of Monthly Charges by Churn Status')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Density')
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.show()

# 9. Churn by Online Services
plt.figure(figsize=(12, 8))
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

plt.subplot(2, 3, 1)
for service in services:
    service_churn = df.groupby(service)['Churn'].mean()
    plt.figure(figsize=(10, 6))
    service_churn.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
    plt.title(f'Churn Rate by {service}')
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45)
    plt.show()

# 10. Churn by Gender and Senior Citizen Status
plt.figure(figsize=(12, 6))
gender_senior_churn = df.groupby(['gender', 'SeniorCitizen'])['Churn'].mean().unstack()
gender_senior_churn.plot(kind='bar', color=['#1f77b4', '#ff7f0e'])
plt.title('Churn Rate by Gender and Senior Citizen Status')
plt.ylabel('Churn Rate')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.legend(title='Senior Citizen', labels=['No', 'Yes'])
plt.show()
```
